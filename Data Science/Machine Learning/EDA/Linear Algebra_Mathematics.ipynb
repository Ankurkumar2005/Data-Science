# Linear algebra in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
2x + 3y = 8
5x + 4y = 20
```

```python
# coefficient matrix A
A = np.array([[2,3],[5,7]])
A
```

```python
# constant matrix B
B = np.array([8,20])
B
```

```python
# solve for x
x = np.linalg.solve(A, B)
x
```

# complex equation

```python
A_complex = np.array([[1+2j, 2-1j], [3-1j, 4]])
B_complex = np.array([4+1j, 3+4j])

X_complex = np.linalg.solve(A_complex, B_complex)
X_complex
```

# Vectors in python

```python
v = np.array([2,3])
v
```

```python
# plot the vector v
plt.figure()
plt.quiver(0,0,v[0],v[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
plt.title('2D vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# Linear Transformation

```python
\[
R = \begin{bmatrix}
\cos(45^\circ) & -\sin(45^\circ) \\
\sin(45^\circ) & \cos(45^\circ)
\end{bmatrix}
\]
```

```python
theta = np.radians(45)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

v_transformed = np.dot(rotation_matrix, v)

plt.figure()
plt.quiver(0,0,v[0],v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Orginal Vector')
plt.quiver(0,0,v_transformed[0],v_transformed[1], angles='xy', scale_units='xy', scale=1, color='b', label='Transformation')
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid()
plt.title('2D vector Transformation')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# Sher Transformation

```python
shear_factor = 2.5
shear_matrix = np.array([[1, shear_factor], [0,1]])

v_sheared = np.dot(shear_matrix, v)

plt.figure()
plt.quiver(0,0,v[0],v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Orginal Vector')
plt.quiver(0,0,v_transformed[0],v_transformed[1], angles='xy', scale_units='xy', scale=1, color='b', label='Rotated Vector')
plt.quiver(0,0,v_sheared[0],v_sheared[1], angles='xy', scale_units='xy', scale=1, color='g', label='Shered Vector')
plt.xlim(-5,20)
plt.ylim(-5,20)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(y=0,color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.title('2D vector Transformation')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
