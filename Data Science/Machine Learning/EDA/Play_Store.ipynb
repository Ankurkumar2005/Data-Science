# convert all into the bytes

```python
def conv_size(Size):
    if isinstance(Size,str):
        if 'k' in Size:
            return float(Size.replace('k', "")) * 1024
        elif 'M' in Size:
            return float(Size.replace('M', "")) * 1024 * 1024
        elif 'Varies with device' in Size:
            return np.nan
    return Size      
```

# now apply

```python
df['Size'] = df['Size'].apply(conv_size)
```

```python
def convert_size(x):
    if isinstance(x, str):
        if 'M' in x:
            return float(x.replace('M', '')) * 1024 * 1024
        elif 'k' in x:
            return float(x.replace('k', '')) * 1024
        else:
            return None
    return x

df['Size_in_bytes'] = df['Size_in_bytes'].apply(convert_size)
```

```python
df['Size_in_mb'] = df['Size_in_bytes'].apply(lambda x: x / (1024 * 1024))
```

# Remove Installs columns
Remove + sign
Remove , 
Convert the columns into integer

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace("+", "") if('+' in str(x)) else(x))
```

```python
df['installs'] = df['Installs'].apply(lambda x : x.replace(',',"") if (',' in str(x)) else(x))
```

```python
df['Installs'] = df['Installs'].apply(lambda x : x.replace('Free',"") if('Free' in str(x)) else(x))
```

```python
df['installs'] = df['Installs'].apply(
    lambda x: int(x.replace(',', '').replace('+', '')) 
    if isinstance(x, str) 
    and x.replace(',', '').replace('+', '').isdigit()
    else 0
)
```

# Reviews

```python
df['Reviews'] = df['Reviews'].apply(
    lambda x: int(float(x.replace('M', '')) * 1_000_000) if isinstance(x, str) and 'M' in x
    else int(float(x.replace('k', '')) * 1_000) if isinstance(x, str) and 'k' in x
    else int(x) if str(x).isdigit()
    else 0
)
```
