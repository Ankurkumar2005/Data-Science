```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'date': ['2001-12-01', '01-12-2022', '2022/12/01', '12-01-2021'],
    'country' : ['USA', 'U.S.A', 'America', 'United States'],
    'name': ['Johan Doe', 'John Doe', 'Jane Doe', 'Jane Doe'],
    'sales_2020': [100,200,None,200],
    'sales_2021':[None, 150,300,150]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

# Date

```python
# Standarizing the date format
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.head()
```

```python
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df.head()
```

# Assigment 
     Fill the date 

```python
df["date"] = df["date"].fillna(method="ffill")
df.head()
```

# country

```python
# Harmonize the name of the country
country_mapping = {'USA': 'United States', 'U.S.A': 'United States', 'America': 'United States'}
df['country'] =  df['country'].replace(country_mapping)
df.head()
```

# Name

```python
# correct the typogrophical Mistake in name
df['name'] = df['name'].replace({'Johan Doe': 'John Doe'})
df.head()
```

```python
# remove duplicates 
df = df.drop_duplicates(subset="name")
df.head()
```
