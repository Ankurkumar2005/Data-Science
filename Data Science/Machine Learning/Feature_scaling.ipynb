```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Feature Scaling

## 1.Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

# Sample Data
data = {'value': [10,20,30,40,50]}
df = pd.DataFrame(data)

# Min-Max Scaling
scaler = MinMaxScaler()
df['scaled_value'] = scaler.fit_transform(df[['value']])
print(df)
```

## 2.Standard Scaler or Z-Score Normalization

```python
from sklearn.preprocessing import StandardScaler

# Sample Data
data = {'value': [10,20,30,40,50]}
df = pd.DataFrame(data)

# Standard Scaler or Z-Score Normalization
scaler = StandardScaler()
df['scaled_value'] = scaler.fit_transform(df[['value']])
print(df)
```

## 3.Robust Scaling

```python
from sklearn.preprocessing import RobustScaler

# Sample Data
data = {'value': [10,20,30,40,50]}
df = pd.DataFrame(data)

# Robust Scaling
scaler = RobustScaler()
df['scaled_value'] = scaler.fit_transform(df[['value']])
print(df)
```

## 4.Logrithmic

```python
# Sample Data
data = {'value': [10000,20000,30000,40000,50000]}
df = pd.DataFrame(data)

# Logrithmic Scaling
scaler = StandardScaler()
df['scaled_value_log'] = np.log(df['value'])
df['scaled_value_log10'] = np.log10(df['value'])
df['scaled_value_log2'] = np.log2(df['value'])

df.head()
```
